```python
#importing modules
import numpy as np
import matplotlib.pyplot as plt
```

```python
#loading csv files into np arraysl
acc_hits = np.loadtxt("https://raw.githubusercontent.com/KwameOtooAppiah/Neuro265/main/acc_hits_hw6.csv", delimiter = ",")
acc_fa = np.loadtxt("https://raw.githubusercontent.com/KwameOtooAppiah/Neuro265/main/acc_fa_hw6.csv", delimiter = ",")
lc_hits = np.loadtxt("https://raw.githubusercontent.com/KwameOtooAppiah/Neuro265/main/lc_hits_hw6.csv", delimiter = ",")
lc_fa = np.loadtxt("https://raw.githubusercontent.com/KwameOtooAppiah/Neuro265/main/lc_fa_hw6.csv", delimiter = ",")
```

```python
# computign Coherence of ACC and LC Hits

# Compute the Fourier transforms
xf = np.array([np.fft.rfft(x - x.mean()) for x in acc_hits])  # ... for each trial in E1
yf = np.array([np.fft.rfft(y - y.mean()) for y in lc_hits])  # ... and each trial in E2

# Compute the spectra
T = time[-1]
dt = time[1] - time[0]
N = 8001
scale = 2 * dt**2 / T
Sxx = scale * (xf * xf.conj()).mean(0)  # Spectrum of E1 trials
Syy = scale * (yf * yf.conj()).mean(0)  # ... and E2 trials
Sxy = scale * (xf * yf.conj()).mean(0)  # ... and the cross spectrum

# Compute the coherence.
cohr = np.abs(Sxy) / (np.sqrt(Sxx) * np.sqrt(Syy))

f = np.fft.rfftfreq(N, dt)                     # Define a frequency axis.
plt.plot(f, cohr.real)                      # Plot coherence vs frequency,
plt.xlim([0, 50])                           # ... in a chosen frequency range,
plt.ylim([0, 1])                            # ... with y-axis scaled,
plt.xlabel('Frequency [Hz]')                # ... and with axes labeled.
plt.ylabel('Coherence')
plt.title('Coherence between ACC and LC Hits')
```

####Analysis of Coherence Plot of ACC and LC Hits

The highest levels of coherence apear to be at the 1-2 hz oscleations and the 6-9 hz osclealations equally. From there the coherence seems to gradually decrease yet is still overal very present among the majority of frequencies.  



```python
# computign Coherence of ACC and LC False Attempts with AAC and LC Hits

# Compute the Fourier transforms
xf = np.array([np.fft.rfft(x - x.mean()) for x in acc_fa])  # ... for each trial in E1
yf = np.array([np.fft.rfft(y - y.mean()) for y in lc_fa])  # ... and each trial in E2

xf1 = np.array([np.fft.rfft(x - x.mean()) for x in acc_hits])  # ... for each trial in E1
yf1 = np.array([np.fft.rfft(y - y.mean()) for y in lc_hits])  # ... and each trial in E2

# Compute the spectra
T = time[-1]
dt = time[1] - time[0]
N = 8001
scale = 2 * dt**2 / T
Sxx = scale * (xf * xf.conj()).mean(0)  # Spectrum of E1 trials
Syy = scale * (yf * yf.conj()).mean(0)  # ... and E2 trials
Sxy = scale * (xf * yf.conj()).mean(0)  # ... and the cross spectrum

Sxx1 = scale * (xf1 * xf1.conj()).mean(0)  # Spectrum of E1 trials
Syy1 = scale * (yf1 * yf1.conj()).mean(0)  # ... and E2 trials
Sxy1 = scale * (xf1 * yf1.conj()).mean(0)  # ... and the cross spectrum

# Compute the coherence.
cohr = np.abs(Sxy) / (np.sqrt(Sxx) * np.sqrt(Syy))
cohr1 = np.abs(Sxy1) / (np.sqrt(Sxx1) * np.sqrt(Syy1))

f = np.fft.rfftfreq(N, dt)                     # Define a frequency axis.

plt.plot(f, cohr.real, color = "blue")                      # Plot coherence vs frequency,
plt.plot(f, cohr1.real, color = "red")                      # Plot coherence vs frequency,
plt.xlim([0, 50])                           # ... in a chosen frequency range,
plt.ylim([0, 1])                            # ... with y-axis scaled,
plt.xlabel('Frequency [Hz]')                # ... and with axes labeled.
plt.ylabel('Coherence')
plt.title('Coherence between ACC and LC Hits and FA')
```

#####Analysis of Combined Coherence Plot



*   The strongest coherence frequency seems to be the same for both sets of data, around 1-2 hz. However, while the hits data had an equally strong coherence at 6-9 hz oscelations, the false alarm data seems to be lacking this.   
*   Coherence seems to be consistently lower for false alarm than it is for hits.
*   The higher coherence for the hits than the false alarms suggests that the two brain areas: anterior cingulate cortex and the locus coeruleus, may be in communciation with each other, as coherence is usually indicaitive of.
