```python
#importing necessary modules
import numpy as np
import matplotlib.pyplot as plt
```

```python
#uploading the files and putting them into arrays
url1 = "https://raw.githubusercontent.com/KwameOtooAppiah/Neuro265/main/acc_fa_hw5.csv"
url2 = "https://raw.githubusercontent.com/KwameOtooAppiah/Neuro265/main/acc_hits_hw5.csv"

acc_hits = np.loadtxt(url1, delimiter = ",")
acc_fa = np.loadtxt(url2, delimiter = ",")
```

```python
#pritning the shape of each array
print(acc_hits.shape)
print(acc_fa.shape)
```

```python
#creating time variable
time = np.linspace(0,4,len(acc_hits[0]))
```

```python
#determinign the sampling rate
rate = len(acc_hits[0])/4
rate
```

```python
#plotting first trial of acc_hits and acc_fa in separate supltos

fig, axes = plt.subplots(2)

axes[0].plot(time, acc_hits[0])
axes[1].plot(time, acc_fa[0])
plt.xlabel('Time [s]')
plt.ylabel('Voltage [$\mu V$]')
```

####Dominant Oscelation

The dominant oscleation of the LFP differs between the data recorded during the hits and the data during the false alarms. During the hits, the dominant oscleation reaches a voltage of near 20 microvolts. In contrast, the dominant oscelation of the LFP at false alarms is above 100 microvolts.

```python
#ERP for acc_hits
plt.plot(time, acc_hits.mean(0))

#with vertical lines indicating moment of screen touch

plt.vlines(2, -10, 20, 'k', lw=1)
plt.hlines(0, 0, 4, 'k', lw=1)

plt.xlabel('Time [s]')
plt.ylabel('Voltage [$\mu V$]')

```

####Interpretation of ERP graph of Hits

In the ERP of the hit condition local field potentials, there appears to be a consistent pattern of oscelations the preceed and follow the 2 second mark. Yet around the 2 second mark, between 1.75 and 2.25, there appears to be a long rising trend in voltage. This may indicate a difference in neuronal activity between the action of the mouse pressing the screen and the act of the mouse recognizing the signal on the screen or preparing to recognize it again.   

```python
#ERP for acc_fa

plt.plot(time, acc_fa.mean(0))
plt.plot(time, acc_hits.mean(0))

#with vertical lines indicating moment of screen touch

plt.vlines(2, -50, 40, 'k', lw=1)
plt.hlines(0, 0, 4, 'k', lw=1)

plt.xlabel('Time [s]')
plt.ylabel('Voltage [$\mu V$]')

```

#### ERP graph of False Alarms compared to Hits
From 0 to 1.5 seconds, the ERPs of each condition appear to be near inverses of each other, due to the microvolt values appearing to miror each other across a horizontal axis at 0. Yet from 1.5 to 2 seconds, the point of the screen touch, there seems to be a near similar pattern of increasing voltage between the two conditions. Both starting from a negative voltage value and steadily approaching 0 microvolts. There is however, a more dramatic difference in the ERPs after the touch between the two conditions. The graph of the false alarms shows a an oscelation that gets increasingly positive then negative, while the graph of the hit condition shows a more consistent pattern to before the screen touch. This might be due to differneces in brain activity when a mouse realizes it hasn't been rewarded compared to when it recognizes that it has been rewarded.
