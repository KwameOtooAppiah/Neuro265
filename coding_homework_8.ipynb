```python
#importing modules
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
```

```python
#loading data
fear_traces = np.loadtxt("https://raw.githubusercontent.com/KwameOtooAppiah/Neuro265/main/fear_conditioning_traces_hw8.csv", delimiter = ",")
extinction_predictors = np.loadtxt("https://raw.githubusercontent.com/KwameOtooAppiah/Neuro265/main/extinction_predictors_hw8.csv")
shock_index = np.loadtxt("https://raw.githubusercontent.com/hallockh/neur_265/main/data/shock_index.csv", delimiter = ",").astype(int)
tone_index = np.loadtxt("https://raw.githubusercontent.com/hallockh/neur_265/main/data/tone_index.csv", delimiter = ",").astype(int)
```

```python
#separating condiiton and extinction traces
conditioning_traces = fear_traces[0:10458]
extinction_traces = fear_traces[10458:]
```

```python
#Printing size of both arrays
print(np.size(conditioning_traces), np.size(extinction_traces))
#didn't add up to the number of samples sO I was thinking maybe this was what the directions wanted:
print(len(conditioning_traces), len(extinction_traces))
```

```python
#making time variables
time_conditioning = np.linspace(0,1/15*np.shape(conditioning_traces)[0],np.shape(conditioning_traces)[0])
time_extinction = np.linspace(0,1/15*np.shape(extinction_traces)[0],np.shape(extinction_traces)[0])

print(time_conditioning, time_extinction)
```

Relationship between Conditioning/Extinction Traces for First ROI adn Second ROI


*   During Conditioning, the cell seems to exhibit more activity during shocks. This is demonstrated by the overlapping of the black lines (timepoints of shocks) and spikes in the measured fluroesence.   
*   This appears to be consistent between both the first and second ROIs. However, the second ROI seems to be slightly higher than the first. This trend is maintiend when examinign the Extinction traces.

```python
# making predictors variable as was made in 4/10 notebook

#creating shock adn tone arrays
shock_array = np.zeros(10458)
for i in range(10458):
  if i >= (shock_index[0]-15)*15 and i <= (shock_index[0]+30)*15:
    shock_array[i] = 1
  elif i>= (shock_index[1]-15)*15 and i <= (shock_index[1]+30)*15:
    shock_array[i] = 1
  elif i>= (shock_index[2]-15)*15 and i <= (shock_index[2]+30)*15:
    shock_array[i] = 1
  else:
    shock_array[i] = 0

tone_array = np.zeros(10458)
for i in range(10458):
  if i >= (tone_index[0]-15)*15 and i <= (tone_index[0]+30)*15:
    tone_array[i] = 1
  elif i>= (tone_index[1]-15)*15 and i <= (tone_index[1]+30)*15:
    tone_array[i] = 1
  elif i>= (tone_index[2]-15)*15 and i <= (tone_index[2]+30)*15:
    tone_array[i] = 1
  else:
    tone_array[i] = 0


#creating predictor variable with tone onset oclumn
from pandas import DataFrame as df
predictors = df(data={'Intercept': np.ones_like(tone_array), 'tone onset': tone_array})

#adding shock onset column
predictors['shock onset'] = shock_array

predictors
```

```python
# running model for each ROI for conditioning_traces

r_index_conditioning = np.zeros(63)
import statsmodels.api as sm
from statsmodels.genmod.families import Poisson

for i in range(63):
  trace_temp = conditioning_traces[:,i]
  model3 = sm.GLM(trace_temp, predictors, family=Poisson())
  model3_results = model3.fit()
  predicted_trace = model3_results.predict()
  r3 = pearsonr(trace_temp, predicted_trace)
  r_index_conditioning[i] = r3.statistic
```

```python
#runnnign model for each ROI for extinction_traces

r_index_extinction = np.zeros(63)

#extinction_predictors had 1 more value than extinction_traces (so I left the last value out): not confident on the impact this will have on the analysis
for i in range(63):
  trace_temp_1 = extinction_traces[:,i]
  model3_1 = sm.GLM(trace_temp_1, extinction_predictors[:14210], family=Poisson())
  model3_results_1 = model3_1.fit()
  predicted_trace_1 = model3_results_1.predict()
  r3_1 = pearsonr(trace_temp_1, predicted_trace_1)
  r_index_extinction[i] = r3_1.statistic

```

```python
#correlating the r_index_extinction and r_index_conditioning variables

r = pearsonr(r_index_conditioning, r_index_extinction)
r
```

```python
#makign scatter plot with each "r" on an axis

plt.scatter(r_index_conditioning, r_index_extinction)
plt.xlabel("R values for Conditioning")
plt.ylabel("R values for Extinction")
```

####Interpretation of Results

Based on the R value of about .015, and the data on the scatterplot it doesn't seem like the cells have high correlation between the correlation of observed and predicted values of Fluresence activity around shock conditioning and extinction. Therefore, since there doesn't seem to be a correlation, it is not likely that a cell that responds strongly to shock adn tone onset durign conditioning will respond strongly during the extinction session.
