```python
#loading data into arrays
spikes_left_trial1 = np.loadtxt("https://raw.githubusercontent.com/KwameOtooAppiah/Neuro265/main/spikes_left_trial1.csv" ,delimiter = ",")
spikes_right_trial1 = np.loadtxt("https://raw.githubusercontent.com/KwameOtooAppiah/Neuro265/main/spikes_right_trial1.csv" ,delimiter = ",")
spikes_left_trial2 = np.loadtxt("https://raw.githubusercontent.com/KwameOtooAppiah/Neuro265/main/spikes_left_trial2.csv" ,delimiter = ",")
spikes_right_trial2 = np.loadtxt("https://raw.githubusercontent.com/KwameOtooAppiah/Neuro265/main/spikes_right_trial2.csv" ,delimiter = ",")
```

```python
#getting length of each array
print(len(spikes_left_trial1))
print(len(spikes_right_trial1))
print(len(spikes_left_trial2))
print(len(spikes_right_trial2))
#interested in seeing whether we can predict which direction rat will turn based on action potential firing frequency
#these data are timestamps from before the rat decided to turn left or right
```

```python
#determining firing rates
print(len(spikes_left_trial1)/spikes_left_trial1[-1])
print(len(spikes_right_trial1)/spikes_right_trial1[-1])
print(len(spikes_left_trial2)/spikes_left_trial2[-1])
print(len(spikes_right_trial2)/spikes_right_trial2[-1])
```

####Interpretation of Firing Rate

Based on the firing rates there seems to be no consistent trend among right turns nor left turns. In left trial 1 the neuron seems to fire more than in left trial 2. Yet, the oppsoite trend is noticied with the right trials. The first right trial has a low firing rate, while the 2nd right trial is higher.

```python
#making subplots of the spike train data from each trial
fig,ax = plt.subplots(2)
ax[0].plot(spikes_left_trial1, 2*np.ones_like(spikes_left_trial1), ".")
ax[0].plot(spikes_right_trial1, np.ones_like(spikes_right_trial1), ".", color = "purple")
ax[0].set_ylim([0,3])
ax[0].set_xlabel("Timestamps[s]")
ax[0].set_title("Trial 1 Left and Right Turn Spikes")


ax[1].plot(spikes_left_trial2, 2*np.ones_like(spikes_left_trial2), ".")
ax[1].plot(spikes_right_trial2, np.ones_like(spikes_right_trial2), ".", color = "purple")
ax[1].set_ylim([0,3])
ax[1].set_title("Trial 2 Left and Right Turn Spikes")
ax[1].set_xlabel("Timestamps[s]")

plt.tight_layout()
```

###Interpretation of Spike trains


*  **Left**: When examining the Left turn data for both trials, both exhibit qualities of bursting. Both have periods of frequent spiking and periods of quietness inbetween. However Left Trial 1 apperas to have longer intervals of spiking lasting around 5 seconds to shorter periods of quiet around 3 seconds. Inversly, Left Trial 2 seems to have smaller periods of spiking around 3 seconds to longer periods of quietness around 5 seconds. These patterns appear most regularly in the 0-20 second interval period yet deviate slightly for both Trials after 20 seconds.
*   **Right**: Both right trials do not seem to exhibit bursting behavior as the activity in the left trials indicated. Both right trials appear to have more compact firing of neurons across time with small intervals of quietness in between them, evidencing more regular action potential firing. However there is a notable period of quietness between the 18 and 25 second timestamp.  

```python
# calculating interspike intervals for each fo the 4 arrays

ISI_L1 = np.diff(spikes_left_trial1)
ISI_R1 = np.diff(spikes_right_trial1)
ISI_L2 = np.diff(spikes_left_trial2)
ISI_R2 = np.diff(spikes_right_trial2)
```

```python
#creating 4 histogram plots
bins = np.arange(0,2, .01)

plt.title("Left Trial 1")
plt.hist(ISI_L1, bins)
plt.xlabel("ISI[s]")
plt.ylabel("Count")
plt.show()

plt.title("Right Trial 1")
plt.hist(ISI_R1, bins)
plt.xlabel("ISI[s]")
plt.ylabel("Count")
plt.show()

plt.title("Left Trial 2")
plt.hist(ISI_L2, bins)
plt.xlabel("ISI[s]")
plt.ylabel("Count")
plt.show()

plt.title("Right Trial 2")
plt.hist(ISI_R2, bins)
plt.xlabel("ISI[s]")
plt.ylabel("Count")
plt.show()


```

####Histogram Interpretation


*   Each graph is rightly sekewed with long tails. This indicates that in each condition, this neuron exhibits bursting behavior. While there a majority of the ISI values are around 0 to a .25 of a second, indicating a consistency in spiking rate, there are also longer ISIs above .5 second long. This demonstrates bursting behavior: frequent spiking separated by longer periods of quietness.
*  There appears to be a difference in the tails of the ISI graphs that exhibits a unique trend among the left and right conditions. Left conditions of trial 1 and 2 tend to have longer tails in their histograms. These tails reach and *exceed* 1.6 second ISIs more consistnelty than the tails of right trial 1 and 2 which only get as high as 1.5 second ISI. This indicates somewhat of a difference in the length of quitness in the bursting between the left and right conditions.



```python
#calculating firing rate across 30 s interval
#ie. 3 spikes in 10 secs ... 6 spikes in 12 secs
#firing rate for trial 1 in subplot 1, firing rate for trial 2 in 2
#by firing rate is it referring to instantenous rate?

binsize = 1                                                                   # Let's start with 1-second bins
fr_bins = np.arange(0,30,binsize)                                             # Make time bins (in 1 second increments) from 0 to 30 s
fr_leftTrial1 = np.zeros(len(fr_bins)-1)                                             # Initialize our firing rate array
for i in range(len(fr_bins)-1):                                               # Loop through each bin
  spk_index = ( spikes_left_trial1 > fr_bins[i]) & (spikes_left_trial1 < fr_bins[i+1])           # Find timestamp values falling within each bin
  true_values = spikes_left_trial1[spk_index]                                          # Our index is a boolean, so we need to isolate "True" values
  fr_leftTrial1[i] = (len(true_values))/binsize

fr_leftTrial2 = np.zeros(len(fr_bins)-1)                                             # Initialize our firing rate array
for i in range(len(fr_bins)-1):                                               # Loop through each bin
  spk_index = ( spikes_left_trial2 > fr_bins[i]) & (spikes_left_trial2 < fr_bins[i+1])           # Find timestamp values falling within each bin
  true_values = spikes_left_trial2[spk_index]                                          # Our index is a boolean, so we need to isolate "True" values
  fr_leftTrial2[i] = (len(true_values))/binsize

fr_rightTrial1 = np.zeros(len(fr_bins)-1)                                             # Initialize our firing rate array
for i in range(len(fr_bins)-1):                                               # Loop through each bin
  spk_index = ( spikes_right_trial1 > fr_bins[i]) & (spikes_right_trial1 < fr_bins[i+1])           # Find timestamp values falling within each bin
  true_values = spikes_right_trial1[spk_index]                                          # Our index is a boolean, so we need to isolate "True" values
  fr_rightTrial1[i] = (len(true_values))/binsize

fr_rightTrial2 = np.zeros(len(fr_bins)-1)                                             # Initialize our firing rate array
for i in range(len(fr_bins)-1):                                               # Loop through each bin
  spk_index = ( spikes_right_trial2 > fr_bins[i]) & (spikes_right_trial2 < fr_bins[i+1])           # Find timestamp values falling within each bin
  true_values = spikes_right_trial2[spk_index]                                          # Our index is a boolean, so we need to isolate "True" values
  fr_rightTrial2[i] = (len(true_values))/binsize

fig,ax = plt.subplots(2)
ax[0].plot(fr_bins[:-1],fr_leftTrial1, color = "red")
ax[0].plot(fr_bins[:-1],fr_rightTrial1, color = "blue")
ax[0].set_xlabel("Time [s]")
ax[0].set_ylabel("Firing Rate of Trial 1[Hz]")

ax[1].plot(fr_bins[:-1],fr_leftTrial2, color = "red")
ax[1].plot(fr_bins[:-1],fr_rightTrial2, color = "blue")
ax[1].set_xlabel("Time [s]")
ax[1].set_ylabel("Firing Rate of Trial 2[Hz]")

plt.tight_layout()
```

####Firing Rate Graph Interpretations


*   In trial 1, the firing rate of the left and the right-turn firing rates appear to be somewhat inversly related. When the firing rate of the left turn is high, the firing rate of the right-turn is low. This seems to be a trend from about 0 to 7 seconds. The left turn seems to reach higher firing rates than the right turn consistently across the time period. Starting around 13 seconds and after, the firing rate seems to fluctuate at low values between 0 and 2 Hz.
*  In trial 2, the firing rate of the left and right dont seem to follow any type of correlating trend. The right-turn firing rate exhibits the highest values reaching about 15 Hz. This exceeds the highest firing rate observed in trial 1 and in this Trial, it appers that the right-turn has a greater firing rate than the left-turn. However, the left-trend seems to exhibit similar firing rates as it did in Trial 1 (between 5 and 10 Hz).
